```python
#This takes a csv file (that contains unicode) from the current directory, 
#reads it into a dictionary, then passes it to pandas to convert to a dataframe 
#for further pandas type processing

import unicodecsv

with open('enrollments.csv', 'rb') as f:
    reader = unicodecsv.DictReader(f)
    enrollments = list(reader)

import pandas as pd
enrollment_df=pd.DataFrame(enrollments)

#I want to get some summary statistics for the enrollments:
enrollment_df.describe()
```
